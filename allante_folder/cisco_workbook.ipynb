```python
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
df = pd.read_pickle('data.pickle')
```

```python
df.info()
```

```python
df.order_date = pd.to_datetime(df.order_date)
```

```python
df.head()
```

```python
df = df.set_index('order_date').sort_index()
df
```

```python
# creating column weekday name 
df['weekday_name']=df.index.day_name()
```

```python
# creating column weekday number 
df['weekday_number']= df.index.day_of_week
```

```python
# creating column month name 
df['month_name']=df.index.month_name()
```

```python
# creating column month name 
df['quarter']=df.index.quarter
```

```python
df.quarter
```

```python
#create subset of data frame 2018 to 2023
df = df['2018':'2023']
```

```python
# list of column names in vertical view 
list(df.columns.values)
```

## What month's have the lowest and highest sales mean?

```python
# average purchase amount per customer 
df.purchase_amount.mean()
```

## takeaway 
Mean purchase price is $4,139

## What month's have the lowest and highest sales mean?

```python
# groupby month name and list highest purchase amount months
df.groupby('month_name').purchase_amount.mean().sort_values(ascending=False)
```

```python
# Visualization 
df.groupby('month_name').purchase_amount.mean().plot()
```

## takeaway 
- December is the month with the highest mean purchase amount 
- October is the month with the lowest mean purchase amounnt 
- School purchase times are around August and December/ school maintenance during December holiday breaks 

```python
## Which Quarter has the highest/ lowest mean? 
df.groupby('quarter').purchase_amount.mean().plot()
```

```python
# mean sales by Quarter 
df.groupby('quarter').purchase_amount.mean()
```

## takeaway 
- Qtr 3 has highest mean purchase amount
